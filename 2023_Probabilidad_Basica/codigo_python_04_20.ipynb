Recordemos la ecuación general de segundo grado. 
$ax^2 +bx+c=0$ entonces las soluciones $x_1$ y $x_2$ que satisfacen la ecuación anterior están dadas por 
$$
x =  \frac{-b \pm \sqrt{b^2-4ac}}{2a}
$$
cuando $b =0$, tenemos que para que existan soluciones, $c <0.$ Luego, se tiene que 
$$
x =  \frac{ \pm \sqrt{ac}}{a}
$$

```python
a = 5
c = 3
h = a*x**2 - c
sp.solve(sp.Eq(h,0))
```
